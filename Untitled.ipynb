```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

* Some wrong values in Experience as it cannot be negative in real life.

```python
obs = data[data['Experience(Years)'] < 0]
```

* Replacing the negative experience values with the median of the values for the corresponding Educaiotn level. 

```python
exp_median = data[data['Experience(Years)'] >= 0].groupby('Education')['Experience(Years)'].median()
```

```python
exp_median
```

```python
#https://sparkbyexamples.com/pandas/pandas-replace-values-based-on-condition/#:~:text=You%20can%20replace%20values%20of,the%20values%20of%20pandas%20DataFrame.

for Education, med_exp in exp_median.items():
    data.loc[(data['Experience(Years)'] < 0) & (data['Education'] == Education), 'Experience(Years)'] = med_exp
```

```python
data.info()
```

```python
cols = ['Personal Loan', 'Securities Account', 'CD Account', 'Online Banking','CreditCard'] 
data[cols] = data[cols].replace({'Yes': 1, 'No': 0})
```

```python
data  = pd.get_dummies(data, columns = ['Education'], dtype=float)
```

```python
data.head()
```

* Splitting to Test train and then using Target encoding to prevent leakge from test to train datasets.

```python
from category_encoders import TargetEncoder

X = data.drop("Income(Thousands's)", axis=1)
y = data["Income(Thousands's)"]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
```

```python
encoder = TargetEncoder(cols=['Sort Code'])
X_train_en = encoder.fit_transform(X_train, y_train)
```

```python
X_train_en.head()
```

```python
train_data = X_train_en.join(y_train)
```

```python
X_test_en = encoder.fit_transform(X_test, y_test)
```

```python
test_data = X_test_en.join(y_test)
```

* Normalizing Non-encoded features.

```python
train_data.hist(figsize=(25,12));
```

```python
from sklearn.preprocessing import normalize

orig_train = train_data.astype(float)
f = normalize(train_data, norm="l2")
train_data = pd.DataFrame(data=f, columns=train_data.columns)

orig_test = test_data.astype(float)
f = normalize(test_data, norm="l2")
test_data = pd.DataFrame(data=f, columns=test_data.columns)

# Replace target variables with original values.
train_data["Income(Thousands's)"] = orig_train["Income(Thousands's)"].values
test_data["Income(Thousands's)"] = orig_test["Income(Thousands's)"].values
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(train_data.corr()[["Income(Thousands's)"]].sort_values(by="Income(Thousands's)", ascending=False), annot=True, cmap='viridis')
```

```python
X_train, y_train = train_data.drop(["Income(Thousands's)"], axis=1), train_data["Income(Thousands's)"]
X_test, y_test = test_data.drop(["Income(Thousands's)"], axis=1), test_data["Income(Thousands's)"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor

xgb = XGBRegressor()
xgb.fit(X_train_s, y_train)
```

```python
def evaluation(y, predictions):
    mae = mean_absolute_error(y, predictions)
    mse = mean_squared_error(y, predictions)
    rmse = np.sqrt(mean_squared_error(y, predictions))
    r_squared = r2_score(y, predictions)
    return mae, mse, rmse, r_squared
```
